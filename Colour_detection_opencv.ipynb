```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("D:\PYTHON PRACTICE(data_analysis)\colorpic.jpg")# importing the picture
```

```python
plt.figure(figsize = (20,10))
plt.imshow(img)
```

## converting the scale to RGB in opencv

```python
grid_RGB = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (20,10))
plt.imshow(grid_RGB)
```

### Converting the color space to HSV as It attempts to depict the colors as perceived by the human eye

```python
grid_HSV = cv2.cvtColor(grid_RGB , cv2.COLOR_RGB2HSV)
```

```python
lower = np.array([95,150,0])## lower and upper for HSV range of light blue color
upper = np.array([110,255,255])
```

```python
mask = cv2.inRange(grid_HSV , lower , upper)
```

```python
plt.figure(figsize = (20,10))
plt.imshow(mask)
```

## Plotted the mask

```python
res = cv2.bitwise_and(grid_RGB , grid_RGB ,  mask = mask)
```

```python
plt.figure(figsize = (20,10))
plt.imshow(res)
```

### Detected the light blue color in the picture

```python
lower_r = np.array([0,150,0])## lower and upper for HSV range of Red color
upper_r = np.array([10,255,255])
```

```python
mask_r = cv2.inRange(grid_HSV , lower_r , upper_r)
```

```python
plt.figure(figsize = (20,10))
plt.imshow(mask_r)
```

```python
res_r = cv2.bitwise_and(grid_RGB , grid_RGB ,  mask = mask_r)
```

```python
plt.figure(figsize = (20,10))
plt.imshow(res_r)
```
